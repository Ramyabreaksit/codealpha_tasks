```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully!")
```

```python
df = pd.read_csv('Downloads/archive (3)/netflix_titles.csv')
print("Dataset loaded successfully!")
print("Shape:", df.shape)
```

```python
# Chart 1: Movies vs TV Shows count
plt.figure(figsize=(8,5))
sns.countplot(x='type', data=df, palette='Set2')
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Chart 2: Top 10 countries with most content
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='Reds_r')
plt.title('Top 10 Countries with Most Netflix Content')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Chart 3: Content added per year
df['year_added'] = pd.DatetimeIndex(df['date_added']).year
year_count = df['year_added'].value_counts().sort_index()

plt.figure(figsize=(12,6))
sns.lineplot(x=year_count.index, y=year_count.values, color='red', marker='o')
plt.title('Netflix Content Added Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.show()
```

```python
# Chart 4: Top 10 ratings
plt.figure(figsize=(12,6))
rating_count = df['rating'].value_counts().head(10)
sns.barplot(x=rating_count.values, y=rating_count.index, palette='Blues_r')
plt.title('Top 10 Content Ratings on Netflix')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

```python
# Chart 5: Top 10 genres
df['listed_in'].value_counts().head(10).plot(kind='barh', figsize=(12,6), color='green')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Count')
plt.tight_layout()
plt.show()
```
